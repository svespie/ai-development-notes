# NumPy Arrays

Can only contain a single data type. 

Where possible, numpy will attempt to coerce the "best" choice in the given dataset. This may not be what is intended, so it is best to be intentional.

https://numpy.org/doc/stable/reference/generated/numpy.array.html 

```python
import numpy as np
```

## Creating Arrays

### 1D Arrays from Python Lists

```python
python_list = [3,2,5,8,4,9,7,6,1]
array = np.array(python_list)
print(array)
type(array)
```

### 2D Arrays from Python Lists

```python
python_list_of_lists = [[3,2,5],
                        [9,7,1],
                        [4,3,6]]

np.array(python_list_of_lists)
```

### Arrays from Scratch

#### np.zeros()

Creates an np array of the specified shape and populated with zeros.

```python
np.zeros((5,3))
```

#### np.random.random()

Creates an np array of the specified shape and populated with random float values.

```python
np.random.random((2,4))
```

#### np.arange()

Creates a 1D np array with evenly spaced values in a specified range with an optional step.

The start value can be omitted, the stop value may not. If three args are presented, the third value represents step.

Notice that the stop value is not included in the array.

```python
np.arange(-3,4)
```

```python
np.arange(4)
```

```python
np.arange(-3, 4, 3)
```
